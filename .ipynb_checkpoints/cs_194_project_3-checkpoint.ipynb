```python
#class implementing the lenet network
class fashionNet(nn.Module):
    def __init__(self):
        super(fashionNet, self).__init__()
        self.cnn_model = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size = 5), #(N, 1, 28, 28) -> (N, 32, 24, 24)
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride = 2), #(N, 32, 24, 24) -> (N, 32, 12, 12)
            
            nn.Conv2d(32, 32, kernel_size = 5), #(N, 32, 12, 12) -> (N, 32, 8, 8)
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride = 2)) #(N, 32, 8, 8) -> (N, 32, 4, 4)
    
        self.fc_model = nn.Sequential(
            nn.Linear(512, 128), # (N, 512) -> (N, 128)
            nn.ReLU(inplace=True),
            nn.Linear(128, 10)) # (N, 128) -> (N, 10)
            
    def forward(self, x):
        x = self.cnn_model(x)
        x = x.view(x.size(0), -1)
        x = self.fc_model(x)
        return x
```

```python
#increase the batch size
batch_size = 128

#download the data again and set the train, test loader with different batch size
train_data = torchvision.datasets.FashionMNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
# trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle=True)
train_data, val_data = torch.utils.data.random_split(train_data, [int(0.8 * len(train_data)), len(train_data) - int(0.8 * len(train_data))])
# print("len of train_data is {}".format(len(train_data)))
# print("len of val_data is {}".format(len(val_data)))
trainloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
valloader = DataLoader(val_data, batch_size=batch_size, shuffle=True)
testset = torchvision.datasets.FashionMNIST(root='./data', train=False, download=True, transform=transforms.ToTensor())
testloader = DataLoader(testset, batch_size=batch_size, shuffle=False)
```

```python
def train(trainloader, net, criterion, optimizer, device, epoch):
    '''
    Function for training.
    '''
    running_loss = 0.0
    net = net.train()
    cnt = 0
    losses = 0
    # for images, labels in tqdm(trainloader):
    for images, labels in trainloader:
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        output = net(images)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        losses += loss.item()*images.size(0)
        cnt += 1
    # return losses/cnt
    return losses

def test(testloader, net, criterion, device):
    '''
    Function for testing.
    '''
    losses = 0
    cnt = 0
    with torch.no_grad():
        net = net.eval()
        # for images, labels in tqdm(testloader):
        for images, labels in testloader:
            images = images.to(device)
            labels = labels.to(device)
            output = net(images)
            loss = criterion(output, labels)
            losses += loss.item()*images.size(0)
            cnt += 1
    # return (losses/cnt)
    return losses
  
def evaluate(dataloader, net):
  #function to calculate the accuracy
    net = net.eval()
    total, correct = 0, 0
    with torch.no_grad():
        for data in dataloader:
            #get the input and labels from data
            inputs, labels = data
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = net(inputs)
            _, pred = torch.max(outputs.data, 1)
            total += labels.size(0)
            #calculate the accuracy
            correct += (pred == labels).sum().item()
        
    return(100 * correct/total)

def plot_train_valid(train_loss, valid_loss, iteration_num, xticks_range=10):
  iteration = [i for i in range(iteration_num)]
  plt.figure()
  plt.plot(iteration, train_loss)
  plt.plot(iteration, valid_loss)
  plt.legend(loc='best', labels=["training", "validation"])
  plt.xlabel("epoch")
  plt.ylabel("error")
  plt.xticks([i for i in range(iteration_num) if i%xticks_range == 0])
  # plt.savefig("part1_loss.jpg")
  plt.show()

def plot_acc_train_valid(train_acc, valid_acc, iteration_num, xticks_range=10):
  iteration = [i for i in range(iteration_num)]
  plt.figure()
  plt.plot(iteration, train_acc)
  plt.plot(iteration, valid_acc)
  plt.legend(loc='best', labels=["training", "validation"])
  plt.xlabel("epoch")
  plt.ylabel("Accuraccy")
  plt.xticks([i for i in range(iteration_num) if i%xticks_range == 0])
  plt.axhline(y=90, color='grey', ls='--')
  # plt.savefig("part1_acc.jpg")
  plt.show()
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
net = fashionNet().to(device)

loss_fn = nn.CrossEntropyLoss()
opt = optim.Adam(net.parameters(), lr = 0.001, weight_decay = 0.0005)

epoch = 30

name = 'fashion_net'

print('\nStart training')
train_losses = []
valid_losses = []
train_accs = []
valid_accs = []
min_valid_loss = np.inf
for epoch in range(epoch): #TODO decide epochs
    print('-----------------Epoch = %d-----------------' % (epoch+1))
    train_loss = train(trainloader, net, loss_fn, opt, device, epoch+1)

    valid_loss = test(valloader, net, loss_fn, device)
    
    train_accuracy = evaluate(trainloader, net)
    train_accs.append(train_accuracy)
    
    valid_accuracy = evaluate(valloader, net)
    valid_accs.append(valid_accuracy)

    train_loss = train_loss/len(trainloader.sampler)
    valid_loss = valid_loss/len(valloader.sampler)
    
    print('[epoch %d] training loss: %.3f' %
          (epoch, train_loss))
    print("validation loss: {}".format(valid_loss))
    print("validation accuracy: ", valid_accuracy)
    print()

    train_losses.append(train_loss)
    valid_losses.append(valid_loss)

    # save model only when validation loss decreases
    if valid_loss < min_valid_loss:
        print('Validation loss decreased ({:.6f} --> {:.6f}). Saving model ...'.format(min_valid_loss, valid_loss))
        min_valid_loss = valid_loss
        torch.save(net.state_dict(), base +'models/model_{}.pt'.format(name))

# load best model
net.load_state_dict(torch.load(base +'models/model_{}.pt'.format(name)))
loss = test(testloader, net, loss_fn, device)/len(testloader.sampler)
accuracy = evaluate(testloader, net)
print("Final Test Loss: ", loss)
print("Fianl Test Accuracy: ", accuracy)
```

```python
def per_class_acc(dataloader, model):
  confusion_matrix = torch.zeros(nb_classes, nb_classes)
  with torch.no_grad():
      for i, (inputs, classes) in enumerate(dataloader):
          inputs = inputs.to(device)
          classes = classes.to(device)
          outputs = model(inputs)
          _, preds = torch.max(outputs, 1)
          for t, p in zip(classes.view(-1), preds.view(-1)):
                  confusion_matrix[t.long(), p.long()] += 1

  # print(confusion_matrix)
  print(confusion_matrix.diag()/confusion_matrix.sum(1))
```

```python
def plot_filters_single_channel(t):
    t = t.cpu()
    
    #kernels depth * number of kernels
    nplots = t.shape[0]*t.shape[1]
    ncols = 8
    
    nrows = 1 + nplots//ncols
    #convert tensor to numpy image
    npimg = np.array(t.numpy(), np.float32)
    
    count = 0
    fig = plt.figure(figsize=(ncols, nrows))
    
    #looping through all the kernels in each channel
    for i in range(t.shape[0]):
        for j in range(t.shape[1]):
            count += 1
            ax1 = fig.add_subplot(nrows, ncols, count)
            npimg = np.array(t[i, j].numpy(), np.float32)
            npimg = (npimg - np.mean(npimg)) / np.std(npimg)
            npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
            ax1.imshow(npimg)
            # ax1.set_title(str(i) + ',' + str(j))
            ax1.set_title(str(i))
            ax1.axis('off')
            ax1.set_xticklabels([])
            ax1.set_yticklabels([])
    # plt.savefig("part1_kernel.jpg")
    plt.tight_layout()
    plt.show()
```

```python
def plot_weights(model, layer_num):
  
  #extracting the model features at the particular layer number
  # layer = model.features[layer_num]
  layer = model.cnn_model[layer_num]
  # layer = model.fc_model[layer_num]
  
  #checking whether the layer is convolution layer or not 
  if isinstance(layer, nn.Conv2d):
    #getting the weight tensor data
    # weight_tensor = model.features[layer_num].weight.data
    weight_tensor = model.cnn_model[layer_num].weight.data
    # weight_tensor = model.fc_model[layer_num].weight.data
    plot_filters_single_channel(weight_tensor)
  else:
    print("Can only visualize layers which are convolutional")
```

```python
class FacadeDataset(Dataset):
    def __init__(self, flag, dataDir= base + '/starter_set/', data_range=(0, 8), n_class=5, onehot=False):
        self.onehot = onehot
        assert(flag in ['train', 'eval', 'test', 'test_dev', 'kaggle'])
        print("load "+ flag+" dataset start")
        print("    from: %s" % dataDir)
        print("    range: [%d, %d)" % (data_range[0], data_range[1]))
        self.dataset = []
        for i in range(data_range[0], data_range[1]):
            img = Image.open(os.path.join(dataDir,flag,'eecs442_%04d.jpg' % i))

            pngreader = png.Reader(filename=os.path.join(dataDir,flag,'eecs442_%04d.png' % i))
            w,h,row,info = pngreader.read()
            label = np.array(list(row)).astype('uint8')

            # Normalize input image
            img = np.asarray(img).astype("f").transpose(2, 0, 1)/128.0-1.0
            # Convert to n_class-dimensional onehot matrix
            label_ = np.asarray(label)
            label = np.zeros((n_class, img.shape[1], img.shape[2])).astype("i")
            for j in range(n_class):
                label[j, :] = label_ == j
            self.dataset.append((img, label))
        print("load dataset done")

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, index):
        img, label = self.dataset[index]
        label = torch.FloatTensor(label)
        if not self.onehot:
            label = torch.argmax(label, dim=0)
        else:
            label = label.long()

        return torch.FloatTensor(img), torch.LongTensor(label)
```

```python
N_CLASS=5

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.n_class = N_CLASS
            #########################################
            ###        TODO: Add more layers      ###
            #########################################
        self.block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            # nn.ReLU(inplace=True),
            nn.Dropout2d(0.2),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.Dropout2d(0.2),
            nn.ReLU(inplace=True)
        )
        self.pool = nn.MaxPool2d((2, 2))

        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            # nn.ReLU(inplace=True),
            nn.Dropout2d(0.2),
            nn.ReLU(inplace=True)
        )
        
        self.up = nn.UpsamplingNearest2d(scale_factor=2)
        self.block3 = nn.Sequential(
            nn.Conv2d(96, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            # nn.ReLU(inplace=True),
            nn.Dropout2d(0.2),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.Dropout2d(0.2),
            nn.ReLU(inplace=True)
        )
        
        self.classfier = nn.Sequential(
            nn.Dropout2d(0.2),
            nn.Conv2d(32, self.n_class, kernel_size=1)
        )

    def forward(self, x):
        out1 = self.block1(x)
        out_pool1 = self.pool(out1)

        out2 = self.block2(out_pool1)

        out_up = self.up(out2)
        out2 = torch.cat((out_up, out1), dim=1)
        out3 = self.block3(out2)

        out = self.classfier(out3)

        return out


def save_label(label, path):
    '''
    Function for ploting labels.
    '''
    colormap = [
        '#000000',
        '#0080FF',
        '#80FF80',
        '#FF8000',
        '#FF0000',
    ]
    assert(np.max(label)<len(colormap))
    colors = [hex2rgb(color, normalise=False) for color in colormap]
    w = png.Writer(label.shape[1], label.shape[0], palette=colors, bitdepth=4)
    with open(path, 'wb') as f:
        w.write(f, label)

def train(trainloader, net, criterion, optimizer, device, epoch):
    '''
    Function for training.
    '''
    start = time.time()
    running_loss = 0.0
    net = net.train()
    cnt = 0
    losses = 0
    # for images, labels in tqdm(trainloader):
    for images, labels in trainloader:
        images = images.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        output = net(images)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        running_loss = loss.item()
        losses += loss.item()
        cnt += 1
    end = time.time()
    # print('[epoch %d] loss: %.3f elapsed time %.3f sum loss %.3f' %
    #       (epoch, running_loss, end-start, losses/cnt))
    print('[epoch %d] training loss: %.3f elapsed time %.3f' %
          (epoch, losses/cnt, end-start))
    return losses/cnt

def test(testloader, net, criterion, device):
    '''
    Function for testing.
    '''
    losses = 0
    cnt = 0
    with torch.no_grad():
        net = net.eval()
        # for images, labels in tqdm(testloader):
        for images, labels in testloader:
            images = images.to(device)
            labels = labels.to(device)
            output = net(images)
            loss = criterion(output, labels)
            losses += loss.item()
            cnt += 1
    return (losses/cnt)


def cal_AP(testloader, net, criterion, device):
    '''
    Calculate Average Precision
    '''
    losses = 0.
    cnt = 0
    with torch.no_grad():
        net = net.eval()
        preds = [[] for _ in range(5)]
        heatmaps = [[] for _ in range(5)]
        for images, labels in tqdm(testloader):
            images = images.to(device)
            labels = labels.to(device)
            output = net(images).cpu().numpy()
            for c in range(5):
                preds[c].append(output[:, c].reshape(-1))
                heatmaps[c].append(labels[:, c].cpu().numpy().reshape(-1))

        aps = []
        for c in range(5):
            preds[c] = np.concatenate(preds[c])
            heatmaps[c] = np.concatenate(heatmaps[c])
            if heatmaps[c].max() == 0:
                ap = float('nan')
            else:
                ap = ap_score(heatmaps[c], preds[c])
                aps.append(ap)
            print("AP = {}".format(ap))

    # print(losses / cnt)
    print("Average AP over 5 class = {}".format(sum(aps)/5))
    return None


def get_result(testloader, net, device, folder='output_train'):
    result = []
    cnt = 1
    with torch.no_grad():
        net = net.eval()
        cnt = 0
        for images, labels in tqdm(testloader):
            images = images.to(device)
            labels = labels.to(device)
            output = net(images)[0].cpu().numpy()
            c, h, w = output.shape
            assert(c == N_CLASS)
            y = np.zeros((h,w)).astype('uint8')
            for i in range(N_CLASS):
                mask = output[i]>0.5
                y[mask] = i
            # print("labels.cpu().data.numpy() is: {}".format(labels.cpu().data.numpy().shape))
            gt = labels.cpu().data.numpy().squeeze(0).astype('uint8')
            save_label(y, './{}/y{}.png'.format(folder, cnt))
            save_label(gt, './{}/gt{}.png'.format(folder, cnt))
            plt.imsave(
                './{}/x{}.png'.format(folder, cnt),
                ((images[0].cpu().data.numpy()+1)*128).astype(np.uint8).transpose(1,2,0))

            cnt += 1

def plot_train_valid(train_loss, valid_loss, iteration_num, xticks_range=10):
  iteration = [i for i in range(iteration_num)]
  plt.figure()
  plt.plot(iteration, train_loss)
  plt.plot(iteration, valid_loss)
  plt.legend(loc='best', labels=["training", "validation"])
  plt.xlabel("epoch")
  plt.ylabel("error")
  plt.xticks([i for i in range(iteration_num) if i%xticks_range == 0])
  plt.show()
```

```python
### hyperparameter
epoch = 100
learning_rate = 1e-3
w_d = 1e-5
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# TODO change data_range to include all train/evaluation/test data.
# TODO adjust batch_size.
# train_data = FacadeDataset(flag='train', data_range=(0,906), onehot=False)
train_data = FacadeDataset(flag='train', data_range=(0, 906), onehot=False)
train_data, val_data = torch.utils.data.random_split(train_data, [int(0.8 * len(train_data)), len(train_data) - int(0.8 * len(train_data))])
train_loader = DataLoader(train_data, batch_size=4, drop_last=True, shuffle=True)
val_loader = DataLoader(val_data, batch_size=4, drop_last=True, shuffle=True)
# test & ap batch size should be 1, or will be cannot be squeezed to 0 error
test_data = FacadeDataset(flag='test_dev', data_range=(0,114), onehot=False)
test_loader = DataLoader(test_data, batch_size=1, drop_last=True, shuffle=True)
ap_data = FacadeDataset(flag='test_dev', data_range=(0,114), onehot=True)
ap_loader = DataLoader(ap_data, batch_size=1, drop_last=True, shuffle=True)

name = 'best_net'
net = Net().to(device)
criterion = nn.CrossEntropyLoss() #TODO decide loss
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate, weight_decay=w_d)

print('\nStart training')
train_losses = []
valid_losses = []
min_valid_loss = np.inf
for epoch in range(epoch): #TODO decide epochs
    print('-----------------Epoch = %d-----------------' % (epoch+1))
    train_loss = train(train_loader, net, criterion, optimizer, device, epoch+1)
    train_losses.append(train_loss)
    # TODO create your evaluation set, load the evaluation set and test on evaluation set
    valid_loss = test(val_loader, net, criterion, device)
    valid_losses.append(valid_loss)
    print("val loss: {}".format(valid_loss))

    # save model only when validation loss decreases
    if valid_loss < min_valid_loss:
        print('Validation loss decreased ({:.6f} --> {:.6f}). Saving model ...'.format(min_valid_loss, valid_loss))
        min_valid_loss = valid_loss
        torch.save(net.state_dict(), base + '/models/model_{}.pt'.format(name))

# load best model
net.load_state_dict(torch.load(base + '/models/model_{}.pt'.format(name)))
print('\nFinished Training, Testing on test set')
test_loss = test(test_loader, net, criterion, device)
print("test loss: {}".format(test_loss))
print('\nGenerating Unlabeled Result')
result = get_result(test_loader, net, device, folder= base + '/output_test')

cal_AP(ap_loader, net, criterion, device)
```
